# 文件的读取 open,read,readlines,readline,close

```python
f = open("./test.txt","r",encoding="UTF-8")
print(f.read(10))
```

## read()：这个函数会影响到f这个实例的指针，下一次再使用read()类似的方法时，就不会从头开始读取，而是从当前指向的位置开始读取。

```python
f = open("./test.txt","r",encoding="UTF-8")
print(f.read(10))  # 已经将指针移动到最后面了
print(f.readlines())  # 报错(如果没报错那么可能是版本问题，老版本应该会报错)
```

## readlines()：返回一个列表，其中每一行的数据为一个元素。

```python
f = open("./test.txt","r",encoding="UTF-8")
print(f.readlines())
```

## readline()：一次读取一行

```python
f = open("./test.txt","r",encoding="UTF-8")
print(f.readline())
print(f.readline())
print(f.readline())
print(f.readline())
```

## for循环读取文件

```python
for line in open("./test.txt","r"):
    print(line)
# 每一个line临时变量，就记录了文件的一行数据
```

## close()：在每次操作完文件之后都要断开连接
- 如果不调用close，同时程序没有停止运行，那么这个文件将一直被python程序占用

```python
f = open("./test.txt","r",encoding="UTF-8")
print(f.read())
f.close()
```

## with open
- 通过在with open的语句块中对文件进行操作
- 可以在操作完成后自动关闭文件，避免忘调用close方法

```python
with open("./test.txt","r") as f:
    print(f.readline())
```
