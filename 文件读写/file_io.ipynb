### 1. 基础打开文件：open 函数（手动关闭）
- 文本模式（默认）：
- 'r' : 只读（默认），文件必须存在
- 'w' : 覆盖写入，文件不存在则创建，存在则清空原有内容
- 'a' : 追加写入，文件不存在则创建，存在则在末尾追加
- 'r+': 可读可写，文件必须存在
- 'w+': 可读可写，文件不存在则创建，存在则清空
- 'a+': 可读可写，文件不存在则创建，存在则在末尾追加

- 二进制模式（加b）：
- 'rb': 只读二进制，'wb': 覆盖写入二进制，'ab': 追加写入二进制
- 用于图片、视频、压缩包等非文本文件

| 说明：推荐优先使用with上下文管理器来操作文件，它会自动处理文件的关闭，避免资源泄漏，同时在发生异常时也能保证文件被正确关闭。

```python
# 基础打开方式，需要手动调用close()关闭文件，否则可能导致资源泄漏
f = open("test.txt", "r", encoding="utf-8")  # r模式：只读，默认模式
content = f.read()
print(content)
f.close()  # 必须手动关闭
```

### 2. 推荐写法：with 上下文管理器（自动关闭）

```python
# with语句会在代码块结束后自动关闭文件，无需手动处理，是Python官方推荐写法
# 即使代码块内发生异常，也能保证文件被正确关闭，避免资源泄漏
with open("test.txt", "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
# 离开with块后，文件自动关闭
```

### 3. 读取整个文件：read ()

```python
with open("test.txt", "r", encoding="utf-8") as f:
    # 一次性读取整个文件的所有内容，返回字符串
    # 适合小文件，大文件使用会占用大量内存
    content = f.read()
    print(content)
```

### 4. 逐行读取：readline ()

```python
with open("test.txt", "r", encoding="utf-8") as f:
    # 每次读取一行，返回字符串，适合大文件逐行处理
    line1 = f.readline()
    line2 = f.readline()
    print(f"第一行: {line1.strip()}")  # strip()去除行尾的换行符
    print(f"第二行: {line2.strip()}")
```

### 5. 读取所有行：readlines ()

```python
with open("test.txt", "r", encoding="utf-8") as f:
    # 一次性读取所有行，返回列表，每个元素对应一行内容
    lines = f.readlines()
    print(lines)  # 输出: ['第一行\n', '第二行\n', '第三行']
    
    # 遍历所有行
    for line in lines:
        print(line.strip())
```

### 6. 迭代文件对象逐行读取（大文件推荐）

```python
with open("test.txt", "r", encoding="utf-8") as f:
    # 文件对象本身是可迭代的，逐行读取，内存占用极低
    # 适合处理GB级的超大文件，不会一次性把文件加载到内存
    for line in f:
        processed_line = line.strip()
        if processed_line:  # 跳过空行
            print(processed_line)
```

### 7. 覆盖写入文件：w 模式 + write ()

```python
with open("test2.txt", "w", encoding="utf-8") as f:
    # w模式：覆盖写入，若文件不存在则创建，若存在则清空原有内容
    # 注意：write()不会自动添加换行符，需要手动添加\n
    f.write("第一行内容\n")
    f.write("第二行内容\n")
```

### 8. 批量写入行：writelines ()

```python
lines = ["第一行\n", "第二行\n", "第三行\n"]
with open("output.txt", "w", encoding="utf-8") as f:
    # 批量写入一个可迭代对象的内容，同样不会自动加换行
    f.writelines(lines)
```

### 9. 追加写入文件：a 模式

```python
with open("output.txt", "a", encoding="utf-8") as f:
    # a模式：追加写入，在文件末尾添加内容，不会清空原有内容
    # 若文件不存在则自动创建
    f.write("这是追加的新行\n")
```

### 10. 读写模式：r + 模式：可读可写

```python
with open("test.txt", "r+", encoding="utf-8") as f:
    # r+模式：可读可写，文件必须存在，文件指针默认在开头
    content = f.read()  # 先读取原有内容
    f.seek(0, 2)  # 移动指针到文件末尾
    f.write("新增的内容\n")  # 写入新内容
```

### 11. 二进制文件读取：rb 模式：二进制模式

```python
# rb模式：只读二进制模式，适合读取图片、视频、压缩包等非文本文件
with open("test.txt", "rb") as f:
    binary_data = f.read()
    # binary_data是bytes对象，可以用于后续的传输或处理
    print(f"读取到 {len(binary_data)} 字节的二进制数据")
```

### 12. 二进制文件写入：wb 模式

```python
# wb模式：覆盖写入二进制模式
data = b"this is binary data"  # 待写入的bytes对象
with open("binary_file.bin", "wb") as f:
    f.write(data)
```

### 13. 文件指针操作：seek () 移动指针

```python
with open("test.txt", "r", encoding="utf-8") as f:
    # seek(offset, whence)：移动文件指针的位置
    # whence=0：从文件开头计算（默认），1：从当前位置，2：从文件末尾
    f.seek(3)  # 从开头移动3个字符的位置
    print(f.read(1))  # 读取第4个字符
    
    f.seek(0)  # 回到文件开头，重新读取
    print(f.readline())
```

### 14. 文件指针操作：tell () 获取当前指针位置

```python
with open("test.txt", "r", encoding="utf-8") as f:
    print(f.tell())  # 初始位置: 0
    f.read(5)
    print(f.tell())  # 读取5个字符后，指针位置: 5
```

### 15. 截断文件：truncate ()

```python
with open("test.txt", "r+", encoding="utf-8") as f:
    # 截断文件，只保留前N个字节的内容，后面的内容会被删除
    f.truncate(10)  # 只保留前10个字节
```

### 16. 刷新缓冲区：flush ()

```python
with open("test.txt", "w", encoding="utf-8") as f:
    f.write("测试内容")
    # 写入的内容默认会先放在内存缓冲区，调用flush()会立即写入磁盘
    f.flush()
    # 也可以等文件关闭时自动刷新缓冲区
```

### 17. CSV 文件读写：pandas 方式

```python
import pandas as pd
# 读取CSV文件
df = pd.read_csv('test.csv', sep=',')

# 写入CSV文件，index=False 不保存行索引
df.to_csv('output.csv', index=False)
```

### 18. CSV 文件读写：内置 csv 模块

```python
import csv
# 读取CSV文件
with open("data.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)

# 写入CSV文件，newline=""避免出现空行
data = [["name", "age"], ["张三", 25], ["李四", 26]]
with open("output.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```
