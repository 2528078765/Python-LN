# 文件的读取 open,read,readlines,readline,close

```python
f = open("./test.txt","r",encoding="UTF-8")
print(f.read(10))
```

## read()：这个函数会影响到f这个实例的指针，下一次再使用read()类似的方法时，就不会从头开始读取，而是从当前指向的位置开始读取。

```python
f = open("./test.txt","r",encoding="UTF-8")
print(f.read(10))  # 已经将指针移动到最后面了
print(f.readlines())  # 报错(如果没报错那么可能是版本问题，老版本应该会报错)
```

## readlines()：返回一个列表，其中每一行的数据为一个元素。

```python
f = open("./test.txt","r",encoding="UTF-8")
print(f.readlines())
```

## readline()：一次读取一行

```python
f = open("./test.txt","r",encoding="UTF-8")
print(f.readline())
print(f.readline())
print(f.readline())
print(f.readline())
```

## for循环读取文件

```python
for line in open("./test.txt","r"):
    print(line)
# 每一个line临时变量，就记录了文件的一行数据
```

## close()：在每次操作完文件之后都要断开连接
- 如果不调用close，同时程序没有停止运行，那么这个文件将一直被python程序占用

```python
f = open("./test.txt","r",encoding="UTF-8")
print(f.read())
f.close()
```

## with open
- 通过在with open的语句块中对文件进行操作
- 可以在操作完成后自动关闭文件，避免忘调用close方法

```python
with open("./test.txt","r") as f:
    print(f.readline())
```

# 文件写入：w模式
- 如果没有就会创建这个文件
- 如果文件存在，使用"w"模式，会请空原有内容

```python
f = open("./test.txt","w",encoding="UTF-8")
f.write("hello world")
f.close()
```

# 文件写入：a模式 追加
- 如果没有就会创建这个文件
- 如果文件存在，使用"a"模式，会在最后位置添加

```python
f = open("./test2.txt","a",encoding="UTF-8")
f.write("he")
f.close()
```

# 练习：
- 将test_copy.txt中的内容复制到test_paste.txt文件中
    - 要求最后两个字为测试的数据不要

```python
f = open("./test_copy.txt","r",encoding="UTF-8")
f2 = open("./test_paste.txt","w",encoding="UTF-8")
for line in f:
    line = line.replace ("\n","")
    if line.endswith("测试") == False:
        f2.write(line+"\n")
f.close()
f2.close()
```
