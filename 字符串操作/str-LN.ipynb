### 1. 字符串不可变特性说明：以下所有字符串方法均不会修改原字符串（因字符串不可变），所有修改类操作都会返回一个新的字符串对象。

```python
# 字符串为不可变类型，所有"修改"操作均返回新字符串，原字符串保持不变
s = "hello"
# s[0] = "H"  # 直接修改字符会报错: TypeError: 'str' object does not support item assignment
new_s = "hello"  # 生成新字符串，原字符串不受影响
print(id(s))
print(id(new_s))
new_s = s.replace("h", "H")
print(f"原字符串: {s}")      # 输出: 原字符串: hello
print(f"新字符串: {new_s}")  # 输出: 新字符串: Hello
print(id(s))
print(id(new_s))
```

### 2. 获取字符串长度（内置函数）

```python
s = "Hello Python 编程"
# len() 直接读取字符串长度属性，效率为O(1)
print(len(s))  # 输出: 15（包含中文、空格，按Unicode字符计数）
```

### 3. 字符串的类型属性：class

```python
s = "hello"
# 获取对象的类型，字符串对象的类型为str
print(s.__class__)  # 输出: <class 'str'>
```

### 4. 大小写转换：capitalize () 首字母大写

```python
s = "hELLO World"
# 仅字符串首字母转为大写，其余所有字母转为小写
print(s.capitalize())  # 输出: Hello world
```

### 5. 大小写转换：title () 每个单词首字母大写

```python
s = "hELLo world"
# 每个单词的首字母转为大写，其余字母转为小写
print(s.title())  # 输出: Hello World
```

### 6. 大小写转换：upper () 全部转为大写

```python
s = "Hello Python"
# 所有字母字符转为大写，非字母字符不变
print(s.upper())  # 输出: HELLO PYTHON
```

### 7. 大小写转换：lower () 全部转为小写

```python
s = "Hello Python"
# 所有字母字符转为小写，常用于不区分大小写的匹配
print(s.lower())  # 输出: hello python
```

### 8. 大小写转换：swapcase () 大小写互换

```python
s = "Hello Python"
# 大写转小写，小写转大写
print(s.swapcase())  # 输出: hELLO pYTHON
```

### 9. 大小写转换：casefold () 消除大小写（通用小写）

```python
# 比lower()更彻底，支持特殊字符，适合跨语言的大小写不敏感匹配
s = "ß"  # 德语字符，lower()会保留ß，casefold()会转为ss
print(s.lower())     # 输出: ß
print(s.casefold())  # 输出: ss
```

### 10. 查找子串：find () 查找首次出现位置

```python
s = "Hello Python, I love Python"
# 从左查找子串，返回首次出现的索引，未找到返回-1（不报错）
print(s.find("Python"))      # 输出: 6
print(s.find("Python", 10))  # 从索引10开始查找，输出: 21
print(s.find("Java"))        # 未找到，输出: -1
```

### 11. 查找子串：rfind () 从右侧查找首次出现位置

```python
s = "Hello Python, I love Python"
# 从右向左查找子串，返回首次找到的索引，未找到返回-1
print(s.rfind("Python"))  # 输出: 21
```

### 12. 查找子串：index () 查找子串位置（未找到报错）

```python
s = "Hello Python"
# 与find()功能一致，但子串不存在时会抛出ValueError
print(s.index("Python"))  # 输出: 6
# print(s.index("Java"))  # 报错: ValueError: substring not found
```

### 13. 查找子串：rindex () 从右侧查找子串位置（未找到报错）

```python
s = "Hello Python, I love Python"
# 从右向左查找，子串不存在时抛出ValueError
print(s.rindex("Python"))  # 输出: 21
```

### 14. 统计子串出现次数：count ()

```python
s = "Hello Python, I love Python"
# 统计子串在字符串中出现的次数，支持指定查找范围
print(s.count("Python"))      # 输出: 2
print(s.count("o"))          # 输出: 4
print(s.count("Python", 0, 10))  # 仅在0-10索引范围内统计，输出:1
```

### 15. 判断前缀：startswith ()

```python
s = "https://www.example.com"
# 判断字符串是否以指定前缀开头，支持传入多个前缀的元组
print(s.startswith("https"))        # 输出: True
print(s.startswith(("http", "ftp")))  # 匹配多个前缀，输出: True
```

### 16. 判断后缀：endswith ()

```python
filename = "report.pdf"
# 判断字符串是否以指定后缀结尾，常用于文件格式验证
print(filename.endswith(".pdf"))         # 输出: True
print(filename.endswith((".pdf", ".doc")))  # 匹配多个后缀，输出: True
```

### 17. 替换子串：replace ()

```python
s = "Hello Python, I love Python"
# 将old子串替换为new子串，count指定最大替换次数（默认全部替换）
s1 = s.replace("Python", "Java")
print(s1)  # 输出: Hello Java, I love Java
s2 = s.replace("Python", "Java", 1)  # 仅替换第一次出现的
print(s2)  # 输出: Hello Java, I love Python
```

### 18. 批量字符替换：translate () + maketrans ()

```python
# 先创建字符映射表，再批量替换字符，适合多字符同时替换
s = "abc123"
# 创建映射：a→x, b→y, c→z
trans_table = str.maketrans({"a":"x", "b":"y", "c":"z"})
print(s.translate(trans_table))  # 输出: xyz123
```

### 19. 分割字符串：split () 从左分割

```python
s = "a,b,c,d"
# 按指定分隔符拆分字符串，返回列表，maxsplit限制最大分割次数
print(s.split(","))          # 输出: ['a', 'b', 'c', 'd']
print(s.split(",", maxsplit=2))  # 仅分割2次，输出: ['a', 'b', 'c,d']

# 不指定分隔符时，默认按任意空白字符拆分
s2 = "Hello   Python\nI\tlove coding"
print(s2.split())  # 输出: ['Hello', 'Python', 'I', 'love', 'coding']
```

### 20. 分割字符串：rsplit () 从右分割

```python
path = "C:/user/documents/test.txt"
# 从右侧开始分割，适合提取文件名/后缀等场景
parts = path.rsplit("/", maxsplit=1)
print(parts)  # 输出: ['C:/user/documents', 'test.txt']
filename = parts[1]
print(filename)  # 输出: test.txt
```

### 21. 按换符分割：splitlines ()

```python
s = "第一行\n第二行\r第三行\r\n第四行"
# 按换行符拆分多行文本，keepends=True可保留换行符
print(s.splitlines())  # 输出: ['第一行', '第二行', '第三行', '第四行']
print(s.splitlines(keepends=True))  
# 保留换行符，输出: ['第一行\n', '第二行\r', '第三行\r\n', '第四行']
```

### 22. 分割为三元组：partition ()

```python
s = "name=张三"
# 按分隔符分割为(前缀, 分隔符, 后缀)三元组，仅分割一次
print(s.partition("="))  # 输出: ('name', '=', '张三')
```

### 23. 从右分割为三元组：rpartition ()

```python
s = "a=b=c"
# 从右侧分割为三元组，仅分割一次
print(s.rpartition("="))  # 输出: ('a=b', '=', 'c')
```

### 24. 拼接序列为字符串：join ()

```python
# 以当前字符串为分隔符，拼接可迭代对象中的所有字符串元素
lst = ["Hello", "Python", "I", "love", "coding"]
print(" ".join(lst))  # 用空格拼接，输出: Hello Python I love coding
print(','.join(lst))  # 用逗号拼接，输出: Hello,Python,I,love,coding
# 批量拼接大量字符串时，join()比+运算符效率高得多
```

### 25. 去除首尾字符：strip ()

```python
s = "  ###Hello Python###  \n"
# 去除首尾的指定字符，默认去除空白字符（空格、换行、制表符）
print(s.strip())          # 去除首尾空白，输出: ###Hello Python###
print(s.strip("# \n"))    # 去除首尾的#、空格、换行，输出: Hello Python
```

### 26. 去除左侧字符：lstrip ()

```python
s = "###Hello Python###"
# 仅去除字符串左侧的指定字符
print(s.lstrip("#"))  # 输出: Hello Python###
```

### 27. 去除右侧字符：rstrip ()

```python
s = "###Hello Python###"
# 仅去除字符串右侧的指定字符
print(s.rstrip("#"))  # 输出: ###Hello Python
```

### 28. 左对齐填充：ljust ()

```python
s = "123"
# 左对齐，右侧用指定字符填充至指定总长度
print(s.ljust(5, "-"))  # 输出: 123--
```

### 29. 右对齐填充：rjust ()

```python
s = "123"
# 右对齐，左侧用指定字符填充至指定总长度
print(s.rjust(5, "-"))  # 输出: --123
```

### 30. 居中对齐填充：center ()

```python
s = "123"
# 居中对齐，两侧用指定字符填充至指定总长度
print(s.center(5, "-"))  # 输出: -123-
```

### 31. 数字补零：zfill ()

```python
s = "123"
# 右侧对齐，左侧填充0至指定长度，适合数字字符串补零
print(s.zfill(5))    # 输出: 00123
print("-123".zfill(5)) # 支持正负号，输出: -0123
```

### 32. 制表符转空格：expandtabs ()

```python
s = "a\tb\tc"
# 将制表符\t替换为指定个数的空格，默认tabsize=8
print(s.expandtabs(4))  # 输出: a   b   c（每个\t替换为4个空格）
```

### 33. 判断是否全为字母：isalpha ()

```python
# 所有字符都是字母（含中文、日文等语言字符），且非空，返回True
print("HelloPython".isalpha())  # 输出: True
print("Hello123".isalpha())    # 输出: False（包含数字）
print("你好".isalpha())         # 输出: True（中文视为字母）
```

### 34. 判断是否全为数字：isdigit ()

```python
# 所有字符都是十进制数字，返回True
print("123456".isdigit())  # 输出: True
print("123.45".isdigit())  # 输出: False（包含小数点）
print("１２３".isdigit())  # 输出: True（全角数字也会识别）
```

### 35. 判断是否全为数字 / 字母：isalnum ()

```python
# 所有字符都是字母或数字，返回True
print("Python123".isalnum())  # 输出: True
print("Python@123".isalnum()) # 输出: False（包含@符号）
```

### 36. 判断是否全为空白字符：isspace ()

```python
# 所有字符都是空白字符（空格、换行、制表符等），返回True
print("   \n\t".isspace())  # 输出: True
print(" a ".isspace())      # 输出: False（包含字母a）
```

### 37. 判断是否为标题格式：istitle ()

```python
# 判断是否每个单词首字母大写，其余字母小写
print("Hello World".istitle())  # 输出: True
print("Hello world".istitle())  # 输出: False
```

### 38. 判断是否全为大写：isupper ()

```python
# 所有字母字符都是大写，返回True，非字母不影响判断
print("HELLO 123".isupper())  # 输出: True
print("Hello".isupper())      # 输出: False
```

### 39. 判断是否全为小写：islower ()

```python
# 所有字母字符都是小写，返回True，非字母不影响判断
print("hello 123".islower())  # 输出: True
print("Hello".islower())      # 输出: False
```

### 40. 判断是否所有字符都是可打印的：isprintable ()

```python
# 判断字符串中所有字符都是可打印的，换行、制表符等不可打印字符会返回False
print("Hello World!".isprintable())  # 输出: True
print("Hello\nWorld".isprintable())  # 输出: False（包含换行符）
```

### 41. 判断是否为 ASCII 字符：isascii ()

```python
# 3.7+新增，判断所有字符都是ASCII字符
print("Hello123".isascii())  # 输出: True
print("你好".isascii())      # 输出: False
```

### 42. 移除前缀：removeprefix () （Python 3.9+）

```python
s = "https://www.example.com"
# 移除字符串开头的指定前缀，若前缀不存在则返回原字符串
print(s.removeprefix("https://"))  # 输出: www.example.com
print(s.removeprefix("http://"))   # 前缀不存在，返回原字符串: https://www.example.com
```

### 43. 移除后缀：removesuffix () （Python 3.9+）

```python
filename = "report.pdf"
# 移除字符串结尾的指定后缀，若后缀不存在则返回原字符串
print(filename.removesuffix(".pdf"))  # 输出: report
print(filename.removesuffix(".doc"))  # 后缀不存在，返回原字符串: report.pdf
```

### 44. 字符串编码：encode ()

```python
s = "Python 编程"
# 将字符串编码为字节串(bytes)，默认编码utf-8
bytes_s = s.encode("utf-8")
print(bytes_s)  # 输出: b'Python\xe7\xbc\x96\xe7\xa8\x8b'

# errors参数处理编码错误，ignore忽略错误，replace替换为�
# s.encode("ascii", errors="ignore")
```

### 45. 字符串格式化：format ()

```python
name = "张三"
age = 25
# 占位符格式化，支持位置、关键字参数，格式控制
s = "我叫{}，今年{}岁".format(name, age)
print(s)  # 输出: 我叫张三，今年25岁

# 进阶格式控制：保留2位小数
salary = 12345.6789
print("月薪：{:.2f}".format(salary))  # 输出: 月薪：12345.68
```

### 46. 字符串格式化：f-string（Python 3.6+ 推荐）

```python
name = "张三"
age = 25
salary = 12345.6789
# 嵌入式格式化，语法更简洁，效率更高
s = f"我叫{name}，今年{age}岁，月薪{salary:.2f}"
print(s)  # 输出: 我叫张三，今年25岁，月薪12345.68
```
