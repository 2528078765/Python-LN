##### 基础聊天调用（一次性返回结果）

```python
import os
from openai import OpenAI

# print(os.environ.get ("DEEPSEEK_API_KEY"))
client = OpenAI(
    api_key=os.environ.get ("DEEPSEEK_API_KEY"),  # 从环境变量读取密钥
    base_url="https://api.deepseek.com"
)
response = client.chat.completions.create (
    model="deepseek-chat",  # 模型名称，不同平台的模型名不同
    messages=[
        # system角色：设定模型的行为、身份
        {"role": "system", "content": "你是一个简洁易懂的Python老师"},
        # user角色：用户的提问
        {"role": "user", "content": "你好"}
    ],
    temperature=0.7,  # 随机性参数，0-2之间，越高回答越随机
    max_tokens=1024,  # 最大生成的token数，控制回答长度
)

# 提取模型返回的回答
print(response.choices[0].message.content)

```

##### 流式调用（逐字返回，实时显示）

```python
import os
from openai import OpenAI

client = OpenAI(
    api_key=os.environ.get ("DEEPSEEK_API_KEY"),  # 从环境变量读取密钥
    base_url="https://api.deepseek.com"
)

# 流式调用，会逐字返回模型的生成结果，适合做聊天界面，模拟打字效果
stream = client.chat.completions.create (
    model="deepseek-chat",
    messages=[
        {"role": "user", "content": "写一个Python的Hello World示例代码"}
    ],
    stream=True,  # 开启流式模式
)

# 遍历流式的返回块，逐字打印
for chunk in stream:
    if chunk.choices[0].delta.content is not None:
        print(chunk.choices[0].delta.content, end="", flush=True)
```

##### 多轮对话的实现（上下文记忆）

```python
import os
from openai import OpenAI

client = OpenAI(
    api_key=os.environ.get ("DEEPSEEK_API_KEY"),  # 从环境变量读取密钥
    base_url="https://api.deepseek.com"
)

# 多轮对话需要把完整的对话历史传给模型，让模型记住上下文
messages = [
    {"role": "system", "content": "你是一个Python老师，回答要简洁"}
]

# 第一轮对话
user_input = "什么是Python的列表？"
messages.append ({"role": "user", "content": user_input})

response = client.chat.completions.create (
    model="deepseek-chat",
    messages=messages
)
assistant_reply = response.choices[0].message.content
print(f"助手: {assistant_reply}")
# 把助手的回复加入消息历史，让下一轮能记住
messages.append ({"role": "assistant", "content": assistant_reply})

# 第二轮对话，模型已经记住了之前的问题
user_input2 = "那它和元组有什么区别？"
messages.append ({"role": "user", "content": user_input2})

response2 = client.chat.completions.create (
    model="deepseek-chat",
    messages=messages
)
print(f"助手: {response2.choices[0].message.content}")

print(f"messages：{messages}")
```

##### 调用的错误处理

```python
# 调用大模型API时，可能会遇到各种错误，需要捕获异常处理
from openai import APIError, APIConnectionError, AuthenticationError

import os
from openai import OpenAI



try:
    client = OpenAI(
        api_key=os.environ.get ("DEEPSEEK_API_KEY"),  # 从环境变量读取密钥
        base_url="https://api.deepseek.com"
    )
    response = client.chat.completions.create (
        model="deepseek-chat",
        messages=[{"role": "user", "content": "测试调用"}]
    )
    print(response.choices[0].message.content)
except AuthenticationError:
    print("错误：API Key无效，请检查你的密钥是否正确")
except APIConnectionError:
    print("错误：网络连接失败，请检查网络或代理设置")
except APIError as e:
    print(f"错误：API调用失败: {e}")
```

##### 无限对话

```python
import os
from openai import OpenAI
from openai import APIError, APIConnectionError, AuthenticationError

try:
    client = OpenAI(
        api_key=os.environ.get("DEEPSEEK_API_KEY"),
        base_url="https://api.deepseek.com"
    )
    # 多轮对话需要把完整的对话历史传给模型，让模型记住上下文
    print("123123123")
    system_input = input("系统Prompt：")
    messages = [
        {"role": "system", "content": system_input}
    ]
    while(True):
        user_input = input()
        if user_input.upper().__eq__("N"):
            break
        messages.append({"role": "user", "content": user_input})
        response = client.chat.completions.create(
        model="deepseek-chat",
        messages = messages
        )
        assistant_reply = response.choices[0].message.content
        print("AI："+assistant_reply)

        


except AuthenticationError:
    print("错误：API Key无效，请检查你的密钥是否正确")
except APIConnectionError:
    print("错误：网络连接失败，请检查网络或代理设置")
except APIError as e:
    print(f"错误：API调用失败: {e}")
```
